```python
#Overall distribution for all players in the dataframe
plt.figure(figsize = (10, 8))
plt.hist(df_player.Overall, bins = np.arange(40, 100, 2), color = "b", rwidth=0.95)
plt.title("Overall Distribution")
plt.xlabel("Overall")
plt.ylabel("Number of Players")
```

```python
#Overall vs Age distribution
plt.figure(figsize = (10, 8))
plt.scatter(df_player.Overall, df_player.Age)
plt.title("Overall vs Age")
plt.xlabel("Overall")
plt.ylabel("Age")
```

```python
#Best young players -> Players with overall greater than or equal to 85 and Age less than or equal to 25
df_player[np.logical_and(df_player["Overall"] >= 85, df_player["Age"] <= 25)]
```

```python
countries_overall = df_player.groupby("NationalTeam")["Overall"].mean()
countries_overall
```

```python
#Overalls of all national teams
plt.figure(figsize = (8,6))
countries_overall.plot(kind = "bar")
plt.title("Highest Rated National Teams")
plt.ylabel("Overall")
```

```python
#Players with highest value 
df_player_val = df_player.sort_values("ValueEUR", ascending = False).head(10)
df_player_val
```

```python
plt.figure(figsize = (10,8))
plt.bar(df_player_val.Name, df_player_val.ValueEUR, color = "g")
plt.title("Highest valued players")
plt.xlabel("Players")
plt.ylabel("Value in BillionEUR")
plt.xticks(rotation = 50)
```

```python
#Players with the highest release clauses
df_player_release = df_player.sort_values("ReleaseClause", ascending = False).head(10)
df_player_release
```

```python
#Players with the highest wages 
df_player_wage = df_player.sort_values("WageEUR", ascending = False).head(10)
df_player_wage
```

```python
#Key attributes for players in attacking, midfield and defending positions
attributes = df_player[['Name', 'Age','Nationality', 'Club','ClubPosition','Overall','Potential','PaceTotal',
       'ShootingTotal', 'PassingTotal', 'DribblingTotal', 'DefendingTotal',
       'PhysicalityTotal', 'Crossing', 'Finishing', 'HeadingAccuracy',
       'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy',
       'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility',
       'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength',
       'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision',
       'Penalties', 'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'STRating', 'LWRating', 'LFRating',
       'CFRating', 'RFRating', 'RWRating', 'CAMRating', 'LMRating', 'CMRating', 'RMRating', 'LWBRating', 'CDMRating',
       'RWBRating', 'LBRating', 'CBRating', 'RBRating']]
attributes
```

```python
#Details for key attributes of attackers
attributes[["PaceTotal", "ShootingTotal", "PassingTotal", "DribblingTotal"]].describe()
```

```python
#Details for specific positional attributes of attackers
attributes[["STRating", "RWRating", "LWRating", "CMRating", "RMRating", "LMRating", "CAMRating"]].describe()
```

```python
#Details for key attributes of defenders
attributes[["DefendingTotal", "PhysicalityTotal"]].describe()
```

```python
#Details for specific positional attributes of defenders
attributes[["LBRating", "RBRating", "CBRating", "LWBRating", "RWBRating"]].describe()
```

```python
#Attackers -> Players with best finishing, ball control, dribbling, sprint speed and potential
attributes.query('Finishing > 85 and BallControl > 85 and Dribbling > 85 and SprintSpeed > 85 and Potential > 80')
```

```python
#Best young strikers -> Players younger than or exactly 25yo with over 82 Striker rating and over 90 potential 
attributes.query('STRating > 82 and Age <= 25 and Potential > 90')
```

```python
#Best young wingers -> Players younger than or exactly 25yo with 85 or over right wing and left wing rating 
#and over 90 potential 
attributes.query('RWRating >= 85 and LWRating >= 85 and Age <= 25 and Potential > 90')
```

```python
#Best young midfielders -> Players younger than or exactly 25yo with 85 or over central, right and left midfield 
#rating and over 90 potential 
attributes.query('CMRating >= 85 and LMRating >= 85 and RMRating >= 85 and Age <= 25 and Potential > 90')
```

```python
#Defenders -> Players with best strength, marking, standing tackle, sliding tackle and interception
attributes.query('Strength > 85 and Marking > 85 and StandingTackle > 85 and SlidingTackle > 85 and Interceptions > 80')
```

```python
#Best young defenders -> Players younger than or exactly 25yo with over 80 central, left and right back rating along with over 
#80 right wing back and left wing back rating and over 85 potential 
attributes.query('CBRating > 80 and LBRating > 80 and RBRating > 80 and LWBRating > 80 and RWBRating > 80 and Age <= 25 and Potential > 85')
```

```python
#Goal Scorers -> Players with best shot power, penalties, free kick accuracy and volleys
attributes.query('ShotPower > 80 and Penalties > 85 and FKAccuracy > 80 and Volleys > 80')
```
